### https://leetcode.com/problems/add-two-numbers/

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        
        r = ListNode()
        result = r
        
        while l1 != None and l2 != None:
            
            if l1.val + l2.val + r.val >= 10:
                r.val = (l1.val + l2.val + r.val) % 10
                r.next = ListNode(val = 1)
                r = r.next
                
            else:
                r.val += l1.val + l2.val
                if l1.next != None or l2.next != None:
                    r.next = ListNode()
                r = r.next
                
            l1, l2 = l1.next, l2.next
        
        if (l1 != None) ^ (l2 != None):
            l = l1 or l2
            while l != None:
                if l.next != None or l.val + r.val >= 10:
                    r.next = ListNode(val = 1 if l.val + r.val >= 10 else 0)
                r.val = (r.val + l.val) % 10
                l, r = l.next, r.next
            
        return result
```

### https://leetcode.com/problems/longest-common-prefix/

```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        
        mlen = min(len(l) for l in strs)
        pref = []
        for i in range(mlen):
            b = True
            for j in range(1,len(strs)):
                if strs[j - 1][i] != strs[j][i]:
                    b = False
                    break
            if b == True:
                pref.append(strs[0][i])
            else:
                break
        return(''.join(pref))
```

### https://leetcode.com/problems/di-string-match/

```python
class Solution:
    def diStringMatch(self, s: str) -> List[int]:
        
        l, r = 0, len(s)
        perm = [None]*(r + 1)
        
        for i in range(len(s)):
            if s[i] == 'D':
                perm[i] = r
                r -= 1
            else:
                perm[i] = l
                l += 1
        perm[i + 1] = l
        
        return perm
```
